```python
fakejob = pd.read_csv('fake_job_postings.csv')
fakejob.head()
```

```python
fakejob.fillna('notgiven',inplace=True)
fakejob
```

```python
# function to get probability of efeatures

def get_prob(X,y):
    # vectorize
    vector = CountVectorizer(max_features=10000)
    vector.fit(X)
    trainX1, testX1,TrainY1,testY1 = train_test_split(X, y, test_size=0.4, random_state=20)
    # transform
    trainX1_abs = vector.transform(trainX1)
    testX1_abs = vector.transform(testX1)
    X1_abs = vector.transform(X)
    # fit to model
    model1 = LogisticRegression(C = 10, n_jobs=-1)
    model1.fit(trainX1_abs, TrainY1)
    # predict score
    testX1_preds = model1.predict(testX1_abs)
    print("score: ", f1_score(testY1, testX1_preds, average='micro'))
    print("confusion matrix: \n\n", confusion_matrix(testY1,testX1_preds))
    # predict probabilities and make a dataframe of 1's probability
    all_preds = model1.predict_proba(X1_abs)
    all_preds_df = pd.DataFrame(all_preds, columns=['zero','one'])
    
    
    return all_preds_df['one']
#     return clf
    
y = fakejob['fraudulent']
```

```python
# i have a better idea... lets check for each of the categorical columns first...
each_fakejob = []
for col in fakejob.iloc[:,:18]:
    if fakejob[col].dtype == 'object':
        print("\n-----------",col,"-----------")
        each_fakejob.append(get_prob(fakejob[col],y))
```
